```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Make sure all object-like or mixed columns are treated as string
for col in X_train.select_dtypes(include=['object', 'category']).columns:
    X_train[col] = X_train[col].astype(str)
    X_test[col] = X_test[col].astype(str)

# Identify categorical columns
categorical_cols = X_train.select_dtypes(include=['object']).columns

# Encode categorical columns with OneHotEncoder
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough'  # keep numerical columns as-is
)

# Fit/transform the training data and transform the test data
X_train_encoded = preprocessor.fit_transform(X_train)
X_test_encoded = preprocessor.transform(X_test)

# Scale the encoded data
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_encoded)
X_test_scaled = scaler.transform(X_test_encoded)

# Train and evaluate KNN for multiple k values
K = []
training = []
test = []
scores = {}

for k in range(2, 25):
    clf = KNeighborsClassifier(n_neighbors=k)
    clf.fit(X_train_scaled, Y_train)

    training_score = clf.score(X_train_scaled, Y_train)
    test_score = clf.score(X_test_scaled, Y_test)

    K.append(k)
    training.append(training_score)
    test.append(test_score)
    scores[k] = [training_score, test_score]

# Plot accuracy
plt.plot(K, training, label='Training Accuracy')
plt.plot(K, test, label='Test Accuracy')
plt.xlabel('K')
plt.ylabel('Accuracy')
plt.title('KNN Accuracy vs. K Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, f1_score, precision_score, recall_score, classification_report, roc_curve, roc_auc_score

# Fit the model
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train_scaled, Y_train)
Y_pred_knn = knn.predict(X_test_scaled)


# Confusion matrix
cm_knn = confusion_matrix(Y_test, Y_pred_knn)

# Error metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(Y_test, Y_pred_knn))
print('Mean Squared Error:', metrics.mean_squared_error(Y_test, Y_pred_knn))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(Y_test, Y_pred_knn)))

# Classification metrics
accuracy_knn = accuracy_score(Y_test, Y_pred_knn)
f1_knn = f1_score(Y_test, Y_pred_knn, average="macro")
p_knn = precision_score(Y_test, Y_pred_knn, average="macro")
r_knn = recall_score(Y_test, Y_pred_knn, average="macro")

print(f"Accuracy = {100*accuracy_knn:.2f}%")
print(f"Precision = {p_knn:.2f}")
print(f"Recall = {r_knn:.2f}")
print(f"F1 Score = {f1_knn:.2f}")
print(classification_report(Y_test, Y_pred_knn))

# ROC Curve (binary only)
if len(np.unique(Y_test)) == 2:
    fpr_knn, tpr_knn, _ = roc_curve(Y_test, Y_pred_knn)
    auc_knn = roc_auc_score(Y_test, Y_pred_knn)

    plt.plot(fpr_knn, tpr_knn, label=f"AUC = {auc_knn:.2f}")
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
    plt.title("ROC Curve - KNN")
    plt.legend(loc=4)
    plt.grid(True)
    plt.show()
else:
    print("Skipping ROC curve â not a binary classification problem.")
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import SpectralClustering
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA

# STEP 1: Clean and scale the data
df_cleaned = df.copy()

# Encode string/object columns
for col in df_cleaned.select_dtypes(include='object').columns:
    df_cleaned[col] = LabelEncoder().fit_transform(df_cleaned[col])

# Scale the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_cleaned)

# STEP 2: Apply Spectral Clustering
sc = SpectralClustering(n_clusters=3, affinity='nearest_neighbors', random_state=42)
sc_labels3 = sc.fit_predict(df_scaled)

# STEP 3: Evaluate Clustering
print("Silhouette Score:", silhouette_score(df_scaled, sc_labels3))
print("Calinski-Harabasz Score:", calinski_harabasz_score(df_scaled, sc_labels3))
print("Davies-Bouldin Score:", davies_bouldin_score(df_scaled, sc_labels3))

# STEP 4: Visualize Clusters using PCA
reduced_data = PCA(n_components=2).fit_transform(df_scaled)
pca_df = pd.DataFrame(reduced_data, columns=["pca1", "pca2"])
pca_df["Cluster"] = sc_labels3

sns.scatterplot(data=pca_df, x="pca1", y="pca2", hue="Cluster", palette="Set1")
plt.title("Spectral Clustering (PCA 2D)")
plt.show()

# STEP 5: Risk Score Calculation Function
def calculate_risk(indexes):
    risk = 0
    for i in indexes:
        diagnosed = df.iloc[i, 36]
        curr_mh = df.iloc[i, 35]
        if curr_mh == 2 and diagnosed == 1:
            risk += 7
        elif curr_mh == 2:
            risk += 2
        elif curr_mh == 1 and diagnosed == 1:
            risk += 2
        elif curr_mh == 1:
            risk += 0.5
        elif curr_mh == 0 and diagnosed == 1:
            risk += 0.25
    return risk

# STEP 6: Segment data by cluster
cluster1 = [i for i, label in enumerate(sc_labels3) if label == 0]
cluster2 = [i for i, label in enumerate(sc_labels3) if label == 1]
cluster3 = [i for i, label in enumerate(sc_labels3) if label == 2]

# STEP 7: Calculate Risk Scores
cluster1_risk = calculate_risk(cluster1)
cluster2_risk = calculate_risk(cluster2)
cluster3_risk = calculate_risk(cluster3)

# STEP 8: Output the results
print("\n--- Cluster Risk Scores ---")
print("Cluster 0 Risk Score:", cluster1_risk)
print("Cluster 1 Risk Score:", cluster2_risk)
print("Cluster 2 Risk Score:", cluster3_risk)

print("\n--- Cluster Assignments ---")
print("Cluster 0 Size:", len(cluster1))
print("Cluster 1 Size:", len(cluster2))
print("Cluster 2 Size:", len(cluster3))
```
